This code shows how to follow the link in your scraper.

As you know we must get the following pages link via scraping... But, sometimes getting the "href" attribute is not easy. It could be buried in Java_Script or in a function. In this case you should manually select the html address of next page and change the page number in a for loop.

And an important note: Please check the websites robots.txt page that you would like to scrape. you can see if there are any restrictions about scraping that url.

For Autoscout24.com you can see the robots.txt file simply writing this:
"https://www.autoscout24.com/robots.txt"

So, let's see it in an example below.

```python
# First, Importing necessary libraries
import scrapy
from scrapy.crawler import CrawlerProcess
```

This time instead of BMW X3 model I will scrape "Alpina" model cars. ( It is again a BMW but special production;)  )

```python
# Defining the spider class
class AlpineSpider_MP(scrapy.Spider):
    
    name = "alpinespider_mp"
    page_number = 2
    allowed_domains = ["autoscout24.com"]
    
    start_urls = ["https://www.autoscout24.com/lst/alpine?sort=price&desc=0&ustate=N%2CU&size=20&page=1&atype=C&"]
    
    def parse(self, response):
        
        items = {}
        
        all_alpine_divs = response.css(".cldt-summary-full-item-main")
        
        for alpine_divs in all_alpine_divs:
            
            model = alpine_divs.css(".cldt-summary-makemodel::text").extract_first()
            version = alpine_divs.css(".cldt-summary-version::text").extract_first()
            price = alpine_divs.css(".cldt-price::text").extract_first()
            km = alpine_divs.css(".cldt-summary-vehicle-data > ul > li:nth-child(1)::text").extract_first()
            registration = alpine_divs.css(".cldt-summary-vehicle-data > ul > li:nth-child(2)::text").extract_first()
            hp = alpine_divs.css(".cldt-summary-vehicle-data > ul > li:nth-child(3)::text").extract_first()
            condition = alpine_divs.css(".cldt-summary-vehicle-data > ul > li:nth-child(4)::text").extract_first()
            previous_owner = alpine_divs.css(".cldt-summary-vehicle-data > ul > li:nth-child(5)::text").extract_first()
            gear_type = alpine_divs.css(".cldt-summary-vehicle-data > ul > li:nth-child(6)::text").extract_first()
            fuel_type = alpine_divs.css(".cldt-summary-vehicle-data > ul > li:nth-child(7)::text").extract_first()
            
            items["model"] = model
            items["version"] = version
            items["price"] = price
            items["km"] = km
            items["registration"] = registration
            items["hp"] = hp
            items["condition"] = condition
            items["previous_owner"] = previous_owner
            items["gear_type"] = gear_type
            items["fuel_type"] = fuel_type
    
            yield items
        
            # next page selector
            next_page = 'https://www.autoscout24.com/lst/alpine?sort=price&desc=0&ustate=N\
            %2CU&size=20&page=' + str(AlpineSpider_MP.page_number) + '&atype=C&'
            
            if AlpineSpider_MP.page_number <= 20:
                
                AlpineSpider_MP.page_number += 1
                
                yield response.follow(next_page, callback = self.parse)
                
        
process = CrawlerProcess() 
 
process.crawl(AlpineSpider_MP) 
 
process.start() 
```

In Autoscout24.com you can see max 20 ads in a single page. Above we see in "item_scraped_count" equals to 63 this means we have scraped 4 pages.

As you know you can save your results to an output file using the code below. Happy Scraping!

```python
#save to a Json file
process = CrawlerProcess({'FEED_URI': 'output.json'})
```

```python
#saving to a csv file
process = CrawlerProcess({'FEED_URI': 'output.csv'})

#I do not recommend saving to csv since csv uses comma to separate colums.
#And this could make your scraped date split into unwanted columns.
```
